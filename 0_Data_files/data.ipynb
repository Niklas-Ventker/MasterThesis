## Dataset A - Wind Speed Turkey

https://www.kaggle.com/datasets/berkerisen/wind-turbine-scada-dataset

```python
# Preprocess the rae data and save it

import numpy as np

# Get the data in an numpy array 
data = np.genfromtxt("/Users/niklas/Documents/GitHub/MasterThesis/0_Data_files/raw_data_files/wind_turbine_scada.csv", delimiter=',', skip_header=1)

# Slice the data to only get the wind speed column
wind_speed = data[:, 2]
wind_speed = wind_speed[~np.isnan(wind_speed)]

np.savetxt("/Users/niklas/Documents/GitHub/MasterThesis/0_Data_files/preprocessed_data_files/wind_speed_small.csv", wind_speed, delimiter=",")

print(f"Dataframe wind_speed has {wind_speed.size} rows.")
```

```python
# Print the preprocessed data

import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/niklas/Documents/GitHub/MasterThesis/0_Data_files/preprocessed_data_files/wind_speed_small.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
print(df.size)
print(df.dtypes)
print(df.max())
print(df.min())
print(df.mean())
print(df.std())

plt.figure(figsize=(20, 5))  
plt.plot(df)
plt.title('wind_speed')
plt.xlabel('time')
plt.ylabel('wind_speed')
plt.show()
```

```python
# Replicate the preprocessed data and save it 
 
import numpy as np

# Get the data in an numpy array 
data = np.genfromtxt('/Users/niklas/Documents/GitHub/MasterThesis/0_Data_files/raw_data_files/wind_turbine_scada.csv', delimiter=',', skip_header=1)

# Slice the data to only get the wind speed column
wind_speed = data[:, 2]
wind_speed = wind_speed[~np.isnan(wind_speed)]

# Replication factor
factor = 1300
base_wind_speed = wind_speed

for i in range(factor - 1):
    wind_speed = np.concatenate((wind_speed, base_wind_speed))

np.savetxt("/Users/niklas/Documents/GitHub/MasterThesis/0_Data_files/replicated_preprocessed_data_files/wind_speed_v2.csv", wind_speed, delimiter=",")

print(f"Dataframe wind_speed has {wind_speed.size} rows.")
```
