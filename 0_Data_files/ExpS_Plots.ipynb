# Exponential Smoothing: Experiment Plots 

Experiments conducted in 4 systems 
- Python Numpy
- Julia

### Performance Comparison

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from experiments
data = {            #Basic For Loop                      #Parallelized Vectorized
    'Julia':        [8.852018594741821,     3.8098859786987305],
    'Numpy':        [30.938296914100647,    2.0273778438568115],
    'R':            [2.381353,              1.349854],
    'SystemDS':     [77.451650333,          3.7598280416]
}

# Groups
groups = list(data.keys())
num_groups = len(data)

# Values
values = np.array(list(data.values()))[:, ::-1]  

# Bar Plot
fig, ax = plt.subplots(figsize=(12, 8)) 
index = np.arange(num_groups)
bar_width = 0.2

# Bar Names in the desired order
bar_names = ['Vectorized', 'Basic For Loop']
bars = []
for i in range(values.shape[1]):
    bar = ax.barh(index + i * bar_width, values[:, i], bar_width, label=bar_names[i])
    bars.append(bar)

# Labels, Title and Legend
# ax.set_ylabel('Systems', fontweight='bold')
ax.set_xlabel('Time in Seconds')
ax.set_title('Simple Exponential Smoothing Performance Comparison: \nDataset A with 65,689,000 rows.\n')
ax.set_yticks(index + bar_width)
ax.set_yticklabels(groups)
ax.set_xlim(0, 85)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)

# Show Values for each Bar
for bar_group in bars:
    for bar in bar_group:
            width = bar.get_width()
            ax.annotate('{:.4f}'.format(width),
                        xy=(width, bar.get_y() + bar.get_height() / 2),
                        xytext=(30, 0),
                        textcoords="offset points",
                        ha='center', va='center',
                        rotation=0)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the Experiments
data = {            #Basic For Loop         #Vectorized             #Parallelized For Loop      #Parallelized Vectorozed
    'Numpy':        [15.070680975914001,    1.4763070344924927,     0.49577343463897705,        1.1732815504074097],
    'Julia':        [11.085314512252808,    6.6144092082977295,     9.984675407409668,          5.461846470832825],
    'R':            [1.364029,              1.958074,               2.501423,                   1.734327],
    'SystemDS':     [106.4595365215,        7.2538843,              102.423557958,            6.9307917499]
}

# Groups
groups = list(data.keys())
num_groups = len(data)

# Values
values = np.array(list(data.values()))

# Bar Plot 
fig, ax = plt.subplots()
index = np.arange(num_groups)
bar_width = 0.2

# Bar Names
bar_names = ['Basic For Loop', 'Vectorized', 'Parallelized For Loop', 'Parallelized Vectorized']
bars = []
for i in range(values.shape[1]):
    bar = ax.bar(index + i * bar_width, values[:, i], bar_width, label=bar_names[i])
    bars.append(bar)

# Labels, Title and Ledgend
ax.set_xlabel('Systems', fontweight='bold')
ax.set_ylabel('Time in Seconds')
ax.set_title('Exponential Smoothing Experiment: Energy Production Dataset - Solar Energy\nDataset with 49.064.400 rows.\n')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(groups)
ax.set_ylim(0, 18)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=4)

# Show Values for each Bar
for bar_group in bars:
    for bar in bar_group:
        height = bar.get_height()
        if height > 18:
            ax.annotate('value cut off: >100',
                        xy=(bar.get_x() + bar.get_width() / 2, 18),
                        xytext=(0, -105),
                        textcoords="offset points",
                        ha='center', va='bottom',
                        rotation=90,
                        color='black')
        else:
            ax.annotate('{:.2f}'.format(height),
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 2),
                        textcoords="offset points",
                        ha='center', va='bottom',
                        rotation=90)

plt.show()
```
