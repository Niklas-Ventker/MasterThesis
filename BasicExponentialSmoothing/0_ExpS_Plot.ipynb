## Wind Turbine Dataset: Wind Speed

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the path to your CSV file
file_path = '/Users/niklas/Documents/GitHub/Uni/10_Masterarbeit/data_files/prepocessed_datafiles/wind_speed_small.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

plt.figure(figsize=(20, 5))  
plt.plot(df)
plt.title('wind_speed')
plt.xlabel('time')
plt.ylabel('wind_speed')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from experiments
data = {            #Basic For Loop         #Vectorized             #Parallelized For Loop      #Parallelized Vectorozed
    'Numpy':        [7.859822988510132,     0.6840405464172363,     0.05615413188934326,        0.26382601261138916],
    'Julia':        [5.656255006790161,     2.58907949924469,       4.914914011955261,          2.6522690057754517],
    'R':            [0.001644,              0.001210,               0.023937,                   0.002033],
    'SystemDS':     [52.433315949757364,    3.43331594975,          52.425658145499995,         3.171802879]
}

# Groups
groups = list(data.keys())
num_groups = len(data)

# Values
values = np.array(list(data.values()))

# Bar Plot 
fig, ax = plt.subplots()
index = np.arange(num_groups)
bar_width = 0.2

# Bar Names
bar_names = ['Basic For Loop', 'Vectorized', 'Parallelized For Loop', 'Parallelized Vectorized']
bars = []
for i in range(values.shape[1]):
    bar = ax.bar(index + i * bar_width, values[:, i], bar_width, label=bar_names[i])
    bars.append(bar)

# Labels, Title and Ledgend
ax.set_xlabel('Systems', fontweight='bold')
ax.set_ylabel('Time in Seconds')
ax.set_title('Exponential Smoothing Experiment: \nDataset Wind Turbine - wind_speed with 25.265.000 rows.\n')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(groups)
ax.set_ylim(0, 10)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=4)

# Show Values for each Bar
for bar_group in bars:
    for bar in bar_group:
        height = bar.get_height()
        if height > 10:
            ax.annotate('value cut off: >50',
                        xy=(bar.get_x() + bar.get_width() / 2, 10),
                        xytext=(0, -105),
                        textcoords="offset points",
                        ha='center', va='bottom',
                        rotation=90,
                        color='black')
        else:
            ax.annotate('{:.4f}'.format(height),
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 2),
                        textcoords="offset points",
                        ha='center', va='bottom',
                        rotation=90)

plt.show()
```
